```python
# Import necessary libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import math

# Import scikit-learn modules
from sklearn import datasets
from sklearn import model_selection
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_regression
from sklearn.model_selection import GridSearchCV 
```

```python
Air_Quality=pd.read_excel("C:\Sravan ML\ML_git\Random Forest\AirQualityUCI.xlsx")
Air_Quality
```

```python
Air_Quality['Date']=pd.to_datetime(Air_Quality['Date'])
Air_Quality.info()
```

```python
Air_Quality.isna().sum()
```

```python
x = Air_Quality.drop(['Date', 'Time', 'CO(GT)'], axis=1) 
y = Air_Quality['CO(GT)']
```

```python

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=5)
```

```python

# Initialize the RandomForestRegressor
reg = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the model using the training sets
reg.fit(x_train, y_train)
```

```python
y_pred = reg.predict(x_test)
y_pred.flatten()
```

```python
from sklearn.metrics import mean_absolute_error

# Ensure y_train and y_pred have the same number of samples
y_train = y_train[:len(y_pred)]

# Calculate the mean absolute error
mae = mean_absolute_error(y_train, y_pred)
print("Mean Absolute Error:", mae)
```

```python
# R2
r2 = r2_score(y_test, y_pred)
r2   
```
